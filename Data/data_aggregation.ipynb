```python
import pandas as pd
import os
from datetime import datetime
```

```python
data1 = os.path.join("raw_data/JC-201904-citibike-tripdata.csv")
data2 = os.path.join("raw_data/JC-201905-citibike-tripdata.csv")
data3 = os.path.join("raw_data/JC-201906-citibike-tripdata.csv")
data4 = os.path.join("raw_data/JC-202004-citibike-tripdata.csv")
data5 = os.path.join("raw_data/JC-202005-citibike-tripdata.csv")
data6 = os.path.join("raw_data/JC-202006-citibike-tripdata.csv")

data_201904 = pd.read_csv(data1)
data_201905 = pd.read_csv(data2)
data_201906 = pd.read_csv(data3)
data_202004 = pd.read_csv(data4)
data_202005 = pd.read_csv(data5)
data_202006 = pd.read_csv(data6)
```

```python
# Merge datasets
frames = [data_201904, data_201905, data_201906, data_202004, data_202005, data_202006]
result = pd.concat(frames)

result.head()
```

```python
# Add a column of "age" in addition to the "birth year"
current_year = datetime.now().year

result["age"] = current_year - result["birth year"]

result.head()
```

```python
# Save the cleaned data in csv
result.to_csv("cleaned_data/bike_qtr2_2019to2020.csv", index=False)

data_result = os.path.join("cleaned_data/bike_qtr2_2019to2020.csv")
data_result = pd.read_csv(data_result)

data_result.head()
```

```python
# Merge and create a smaller data for testing purposes
frames1 = [data_201904, data_202004]
result1 = pd.concat(frames1)

result1["age"] = current_year - result1["birth year"]

# Export the small data
result1.to_csv("cleaned_data/bike_april_2019to2020.csv", index=False)

data_result1 = os.path.join("cleaned_data/bike_april_2019to2020.csv")
data_result1 = pd.read_csv(data_result1)

data_result1.count()
```
